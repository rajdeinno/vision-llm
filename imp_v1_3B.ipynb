```python
model = AutoModelForCausalLM.from_pretrained(
    "MILVLG/imp-v1-3b",
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True
)

tokenizer = AutoTokenizer.from_pretrained(
    "MILVLG/imp-v1-3b",
    trust_remote_code=True
)
```

```python
# text = """A chat between a curious user and an artificial intelligence assistant for creating test cases.
# The assistant generates helpful and detailed field names and data types such as textbox, dropdown, labels. I need the result in json format for each fields.
# USER: <image>\n Analyze the form data as JSON. ASSISTANT:"""

text="""A chat between a user and an AI assistant focused on extracting form fields from images. The assistant identifies and extracts different types of form fields such as
textboxes, labels, dropdowns, checkboxes, etc.,
along with their corresponding names or identifiers.
The result should be returned in JSON format, with each field type paired with its name or label.
The field types may include 'textbox', 'label', 'dropdown', 'checkbox', etc., and
the field names are the corresponding headers or identifiers next to the fields. Do not generate your
own field names.

Here are a few examples to understand the expected format:

Example 1:
For example, json
[
  {"textbox": "Transaction Type"},
  {"label": "Create Deal"},
  {"dropdown":"Type"}
]
USER: <image>\n Analyse the image and create the JSON data from the prompt.
ASSISTANT:
"""

image = Image.open("/content/gdrive/MyDrive/gen-ai/vison-llm/images/form2.png")
```

```python
input_ids = tokenizer(text, return_tensors="pt").input_ids
image_tensor = model.image_preprocess(image)
```

```python
output_ids = model.generate(
    input_ids,
    max_new_tokens=300,
    images=image_tensor,
    use_cache=False
)[0]
```

```python
print(tokenizer.decode(output_ids[input_ids.shape[1]:], skip_special_tokens=True).strip())
```

```python
import gradio as gr
```

```python
def generate_answer(image):
  input_ids = tokenizer(text, return_tensors="pt").input_ids
  image_tensor = model.image_preprocess(image)
  output_ids = model.generate(
      input_ids,
      max_new_tokens=300,
      images=image_tensor,
      use_cache=True)[0]
  result = tokenizer.decode(output_ids[input_ids.shape[1]:], skip_special_tokens=True).strip()
  return result
```

```python
# Create a Gradio interface
iface = gr.Interface(fn=generate_answer,
                     inputs=gr.inputs.Image(type="pil"),
                     outputs="text",
                     title="Multimodal Demo app using a Small Vision Model",
                     description="Upload an image to analyze.")

# Run the app
iface.launch(share=True, debug=True)
```
